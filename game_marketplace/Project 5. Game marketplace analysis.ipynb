**Мы не можем заменять пропуски в столбцах с рейтингами и оценками экспертов/пользователей** - они означают либо, что такой оценки ещё нет, либо что её забыли проставить или она "пропала" по какой-то другой причине - в любом случае, проставление оценки нами будет сильно искажать картину, т.к. даже две игры одной серии могут значительно отличаться друг от друга.

**Теоретически, значения "tbd" можно заменить на пропуски** (это будет сделано на этапе статистического анализа гипотез), однако до этого **было бы интересно обратить на них внимание отдельно** (на этапе анализа корреляции с продажами), поэтому трогать эти значения мы пока что не будем.

На данный момент остаётся только столбец `year_of_release`. Давайте посмотрим, как обстоит ситуация с ним.

```python
data[data.year_of_release.isna()].head(10)
```

Видим, что в списке есть игры, год выпуска которых есть непосредственно в их названии. Давайте посмотрим, **сколько всего таких, и что это за игры**

```python
years_names = [*range(1990,2022)]
games_with_years = []

nulls = data[data['year_of_release'].isna()]
nulls.reset_index(inplace=True)
for i in range(len(nulls)):
    for y in years_names:
        if str(y) in nulls.loc[i, 'name']:
            games_with_years.append(nulls.loc[i, 'name'])

games_with_years
```

Заменим пропуски в датах этих игр на даты в названиях:

```python
#Создаём словарь
replacement_order = {}
for y in years_names:
    for game in games_with_years:
        if str(y) in game:
            replacement_order[game] = y-1
replacement_order
```

```python
#Производим замену:
data.loc[data['year_of_release'].isnull(), 'year_of_release'] = data['name'].map(replacement_order)

#Т.к. анализ мы будем проводить по конкретным временным периодам 
#то позиции без указанной даты выпуска нам бесполезны. Их можно удалить.
data.dropna(subset=['year_of_release'], inplace=True)
data.reset_index(inplace=True, drop=True)
```

Причины выбора периода:
- Небольшой кабинетный анализ* показал, что **этот период можно выделить как отдельную веху в развитии рынка** - рост популярности мобильных игр, изменение направления разработки в ряде жанров. (*https://online.maryville.edu/blog/future-of-video-games/ и другие ресурсы)
- Анализ количества выпускаемых игр показало, что **со временем платформы имеют свойство "затухать", в связи с чем рассматривать платформы, которые уже "затухли" или находятся в активной фазе "затухания" не имеет смысла**, т.к. на их основании мы не сможем дать объективного прогноза на следующий год (за исключением анализа "циклов" платформ).

```python
#Картина по количеству выпущенных игр примерно понятна. А теперь давайте посмотрим на число продаж за обозначенный период.
years_to_drop = [*range(1980,2010)]

platform_income_by_year = data.pivot_table(index='platform', columns='year_of_release', values='total_sales', aggfunc='sum')

platform_income_by_year = platform_income_by_year.drop(columns=years_to_drop) #Убираем неинтересующие нас года

platform_income_by_year = platform_income_by_year.assign(sum=platform_income_by_year.sum(axis=1)) #Считаем общую выручку за период

platform_income_by_year = platform_income_by_year.sort_values(by='sum', ascending=False)\
.drop(columns='sum')\
.head(12) #Получено путём просмотра таблицы - именно у этих платформ есть какие-либо продажи в исследуемый период

#Построим ещё один хитмэп
fig, ax = plt.subplots(figsize=(30,20))
sns.heatmap(platform_income_by_year, annot=True, fmt='g', ax=ax, annot_kws={"size":18})
ax.tick_params(labelsize=16)
plt.title("Общее число продаж на каждой игровой платформе игр по годам", weight='bold').set_fontsize(25)
plt.show();
```

```python
#Старая версия матплотлиба - нет функции ax.bar_label
import matplotlib 
print('matplotlib: {}'.format(matplotlib.__version__))
```

```python
#Функция честно сворована и чуть-оптимизирована - https://www.statology.org/seaborn-barplot-show-values/
#Есть более простой способ, но он, как указано выше, требует более новой версии MatPlotLib.PyPlot

def display_values(axs, space=.01):
    def _single(ax):
        for p in ax.patches:
            _x = p.get_x() + p.get_width() / 2
            _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
            value = '{:.1f}'.format(p.get_height())
            ax.text(_x, _y, value, ha="center") 

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
#Давайте выпишем топ-5 платформ по сумме продаж по каждому региону:
for column in data_platforms_regional_sum.columns:
    data_column_top5 = data_platforms_regional_sum.sort_values(by=column, ascending=False).head(5)
    
    fig, ax = plt.subplots()
    ax = sns.barplot(data=data_column_top5, 
            x=data_column_top5.index, 
            y=column,
            palette="Blues_r")
    
    display_values(ax)
    plt.title(column, weight='bold').set_fontsize('18')
    plt.show();
```

```python
#Для начала заменим "tbd" в столбце "user_score" на пропуски и уберём их из анализа.
#По умолчанию Scipy.Stats должен их игнорировать - однако у меня они почему-то ломают показатель (не может провести t-test)
data_top9['user_score'] = data_top9['user_score'].replace('tbd', np.nan)
data_top9['user_score'] = data_top9['user_score'].astype(float)
data_ratings = data_top9.dropna(subset=['user_score'])
```

```python
#Посмотрим, равна ли дисперсия у этих показателей.
xone_var = np.var(data_ratings[data_ratings['platform'] == 'xone']['user_score'])
pc_var = np.var(data_ratings[data_ratings['platform'] == 'pc']['user_score'])

print(f'Значение дисперсии для Xbox One равняется {xone_var:.2f}')
print(f'Значение дисперсии для PC равняется {pc_var:.2f}')
```

```python
#Используем метод "ind", т.к. это независимые друг от друга показатели.
#Параметр "equal_var" ставим "False", т.к. дисперсия здесь не равна

result = st.ttest_ind(data_ratings[data_ratings['platform'] == "xone"]['user_score'],
                     data_ratings[data_ratings['platform'] == 'pc']['user_score'],
                     equal_var = False)

#Задаём значение alpha. Значение подбиралось несколько раз - но из-за крайне малого значения pvalue мы можем отвергнуть гипотезу
#о равенстве средних даже при крайне малом alpha.
alpha = 0.001

#Выведем результат на экран
print(result.pvalue)
if result.pvalue < alpha:
    print("Отвергаем нулевую гипотезу о равенстве средних")
else:
    print("Не можем отвергнуть нулевую гипотезу")
```

```python
result = st.ttest_ind(data_ratings[data_ratings['genre'] == 'action']['user_score'],
                     data_ratings[data_ratings['genre'] == 'sports']['user_score'],
                     equal_var = False)

#Задаём значение alpha. Значение подбиралось несколько раз - но из-за крайне малого значения pvalue мы можем отвергнуть гипотезу
#о равенстве средних даже при крайне малом alpha.
alpha = 0.001

#Выведем результат на экран
print(result.pvalue)
if result.pvalue < alpha:
    print("Отвергаем нулевую гипотезу о равенстве средних")
else:
    print("Не можем отвергнуть нулевую гипотезу")
```
